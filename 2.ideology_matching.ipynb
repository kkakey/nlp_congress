## Categorize Ideology

```python
def ideo_vec(df):
    '''Categorizes ideology into three categories:
    extreme liberal, moderate, extreme conservative.
    Two measurements: ideo_vec, ideo_vec_ext'''
    newdf = pd.DataFrame()
    for session, df in df.groupby('congress'):
        print(session)
        grouped_speeches = df.groupby(['party_R']) #to calculate median cfscores by party
        l = pd.Series.tolist(df[['cfscore','party_R','fullname','speechid']])
        ideo_vec = []
        ideo_vec_ext = []
        count = 0
        for i in l:

            if i[2] not in nan_cfscore:
            
            #ideo_vec
                #conservatives - ext.
                if (i[0] >= grouped_speeches['cfscore'].quantile([.25, .5, .75])[1].iloc[1]):
                    ideo_vec.append(1)

                #mod.
                if (i[0] < grouped_speeches['cfscore'].quantile([.25, .5, .75])[1].iloc[1]) and (i[0]> grouped_speeches['cfscore'].quantile([.25, .5, .75])[0].iloc[1]):
                    ideo_vec.append(0) 

                #liberal - ext.
                if (i[0] <= grouped_speeches['cfscore'].quantile([.25, .5, .75])[0].iloc[1]):
                    ideo_vec.append(-1)

            #ideo_vec_ext
                #conservatives - ext.
                if (i[0] >= grouped_speeches['cfscore'].quantile([.25, .5, .75])[1].iloc[2]):
                    ideo_vec_ext.append(1)

                #mod.
                if (i[0] < grouped_speeches['cfscore'].quantile([.25, .5, .75])[1].iloc[0]) and (i[0]> grouped_speeches['cfscore'].quantile([.25, .5, .75])[0].iloc[2]):
                    ideo_vec_ext.append(0)

                #liberal - ext.
                if (i[0] <= grouped_speeches['cfscore'].quantile([.25, .5, .75])[0].iloc[0]):
                    ideo_vec_ext.append(-1)
                    
                    
                if (i[0] <= grouped_speeches['cfscore'].quantile([.25, .5, .75])[0].iloc[2]) and (i[0]>grouped_speeches['cfscore'].quantile([.25, .5, .75])[0].iloc[0]):
                    ideo_vec_ext.append("NA")

                if (i[0] >= grouped_speeches['cfscore'].quantile([.25, .5, .75])[1].iloc[0]) and (i[0]<grouped_speeches['cfscore'].quantile([.25, .5, .75])[1].iloc[2]):
                    ideo_vec_ext.append("NA")

            if i[2] in nan_cfscore:
                ideo_vec.append("NA")
                ideo_vec_ext.append("NA")

            count += 1
            if count % 2000==0:
                print(count)

        df['ideo_vec']= ideo_vec
        df['ideo_vec_ext']= ideo_vec_ext
        newdf = newdf.append(df, ignore_index=True)
        print('df done')
        
    return(newdf)
```

### Add Summary Statistic Variables

```python
def speech_length(df_c):
    '''count number of words in each speech'''
    l = pd.Series.tolist(df_c[['original speech']])
    ct = 0
    count_per_speech = []
    for line in l:
        for i in line:
            i = i.split()
            ct += 1
            count_per_speech.append(len(i))
    df_c['speech_length'] = count_per_speech
    return(df_c)
```

```python
def summary_stats(df_c):
    '''sent_length = num of sentences in speech
       num_word_unique = num of unique words in speech'''
    from nltk.tokenize import sent_tokenize, word_tokenize
    
    sent_len = []
    unique_len = []
    body_basic = []
    punctuations = '''!()-[]{}';:'"\,<>./?@#$%^&*_~``'''
    ct = 0
    for i in df_c['original speech']:
        sents = sent_tokenize(i)
        tmp = word_tokenize(i)
        tmp_words = [i.lower() for i in tmp if i not in punctuations]
        tmp_unique = len(set(tmp))
        tmp_words = ' '.join(tmp_words)
        sent_len.append(len(sents))
        unique_len.append(tmp_unique)
        body_basic.append(tmp_words)
        
        ct += 1
        if ct % 1000==0:
            print(ct)
            
    df_c['body_basic'] = body_basic
    df_c['sent_length'] = sent_len
    df_c['num_word_unique'] = unique_len
    return(df_c)
```

```python
df_final = speech_length(df_final)
df_final = summary_stats(df_final)
```
